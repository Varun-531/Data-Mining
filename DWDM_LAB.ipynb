```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/students_placement_data.csv')
data
```

```python
X = data.drop('Gender',axis=1)
X = X.drop('Section',axis=1)
X = X.drop('registered_for_ Placement_Training',axis=1)
X = X.drop('Roll No',axis=1)
X
```

# **categorial data**

```python
data['no of backlogs'] = pd.cut(data['Backlogs'],bins=[-1,10,20,np.inf],labels=["0-10","10-20",">20"])
data.head(4)
```

```python
data['Rating'] = pd.cut(data['Backlogs'],bins=[-1,10,20,np.inf],labels = ["Good","Average","Worst"])
data
```

# **Scaling data**

```python
X = X.drop('placement status',axis =1 )
Y = data['placement status']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# **Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_scaled,Y,test_size=0.2)
```
